```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#1.1
data = pd.read_csv('global_air_quality.csv')
data
```

```python
#1.2
data.insert(len(data.columns), 'PM25_Value', data['FactValueNumeric'])
data.loc[0:9450, ['FactValueNumeric', 'PM25_Value']]
```

```python
#1.3
new_data = data.groupby('ParentLocation', as_index=False)['PM25_Value'].mean()
new_data.rename(columns={'PM25_Value': 'Average_PM25'}, inplace=True)
merged_data = data.merge(new_data, on='ParentLocation', how='left')
print(merged_data[['Location', 'ParentLocation', 'PM25_Value', 'Average_PM25']])
```

```python
#1.4
highest = new_data.sort_values('Average_PM25', ascending = False, ignore_index = True)
highest.iloc[0][['ParentLocation', 'Average_PM25']]
# Yes, this surprises me since I feel like I would assume this region would be on the lower end due to fewer cars, or so I had thought. 
# I thought public transportation was much more popular in this area than it seems to be. 
# I also don't know of any factories/products in this area that could produce these emissions. 
```

```python
#1.5
final_data = merged_data[['Location', 'ParentLocation', 'PM25_Value', 'Average_PM25']]
final_data.to_csv('new_global_air_quality.csv', sep=',', index=False)
```

```python
#2.1
planets = sns.load_dataset('planets')
plt.figure(figsize=(20, 20))
sns.scatterplot(planets, x = 'orbital_period', y = 'mass', hue = 'method')
plt.xscale('log')
plt.yscale('log')
plt.title('Orbital Period vs Mass')
plt.xlabel('Planet Orbital Period (Days)')
plt.ylabel('Planet Mass (Earths)')
plt.show()
```
