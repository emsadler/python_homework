```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1
```

```python
data = np.linspace(0, 2*np.pi, 1000)
data
```

```python
amplitude = np.array([1,2,3,4,5])
amplitude
```

```python
wavelength = np.array([0.5,1,1.5,2,2.5])
wavelength
```

```python
A = amplitude
w = wavelength
x = data
for A, w in zip(amplitude, wavelength):
    y = np.sin(w*x)*A
    plt.plot(x,y)

plt.title('5 Different Plots')
plt.xlabel('x data')
plt.ylabel('y data')
```
