```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.optimize as fit
```

```python
data = pd.read_csv('python_project_data.csv')
data
```

```python
data['Vave Squared'] = data['Vave (km/s)'] ** 2
data['r (km)'] = data['r (arcsec)'] / 206265 * 1.496e8
data
```

```python
x = data['r (km)']
v = data['Vave Squared']
G = 6.6743e-11

y = (v/G)*x

plt.figure(figsize=(20, 5))
plt.scatter(x, (v/G)*x)
plt.xlabel('Radius (km)')
plt.ylabel('Mass Predicted')
plt.title('Mass vs Radius')
plt.show()
```

```python
x = data['r (km)']
v = data['Vave Squared']
G = 6.6743e-11

y = (v/G)*x

m, c = np.polyfit(x, y, 1)
y_fit = m*x + c

plt.figure(figsize = (20, 5))
plt.scatter(x, y, label = 'Data')
plt.plot(x, y_fit, label = 'Line of Best Fit', color = 'purple')
plt.xlabel('Radius (km)')
plt.ylabel('Mass Predicted')
plt.title('Mass vs Radius')
plt.legend()
plt.show()
```

```python
x = data['Vave Squared']
r = data['r (km)']
G = 6.6743e-11

y = (r/G)*x

plt.figure(figsize=(20, 5))
plt.scatter(x, y, color = 'red')
plt.xlabel('Velocity Squared')
plt.ylabel('Mass Predicted')
plt.title('Mass vs Velocity Squared')
plt.show()
```

```python
x = data['Vave Squared']
r = data['r (km)']
G = 6.6743e-11

y = (r/G)*x

m, c = np.polyfit(x, y, 1)
y_fit = m*x + c

plt.figure(figsize=(20, 5))
plt.scatter(x, y, label='Data', color='red')
plt.plot(x, y_fit, label = 'Line of Best Fit', color='orange')
plt.xlabel('Velocity Squared')
plt.ylabel('Mass Predicted')
plt.title('Mass vs Velocity Squared')
plt.legend()
plt.show()
```
